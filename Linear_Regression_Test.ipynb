```python
#importing pandas Library 
import pandas as pd

# Step 2: Defining the columns of and reading our DataFrame
#columns is a list of string values
#Fish_Text_File.txt is a txt file that records various informaiton about fish
#FISH is a data frame object?
columns = ['Species', 'Weight',	'V-Length', 'D-Length', 'X-Length', 'Height', 'Width']
Fish = pd.read_csv('Fish_Text_File.txt', sep='\t', usecols = columns)

#Printing the head of our DataFrame
print(Fish.head())
```

```python
#from this we can concluse that fish is likely some sort of string in Python
#ACTUALLY this is wrong, this is a special object called a data frame.
#in other words, Fish is a data frame
print(Fish)
```

```python
#A simple linear regression model uses only one indepdent variable.
#When making a model with only feature of the fish, we would want to use the most strongly predictive
#feature of whatevever we are looking for. 
#One way to do this is to accomplish this with a cross-correlation between the various features

#changed columns such that newColumns now excludes species, which is a string a threfore
#causes an error when looking for the correlation matrix
newColumns = [ 'Weight',	'V-Length', 'D-Length', 'X-Length', 'Height', 'Width']

#Now there is a new set of columns so the newFish data frame should now include only float variables
newFish = pd.read_csv('Fish_Text_File.txt', sep='\t', usecols = newColumns)

#outputting newFish
print(newFish)

#Finding the cross correlation matrix 
print(newFish.corr())

#The correlaiton matrix allows to see which variables are most strongly related to the others
```

```python
#Step 3: separating the data into features(independant variables) and labels(dependent variables)

#In this case we are choosing to use only one featre(X-Length) to predict the label(Weight)
#This is because there is  a strong correlation between X-Length and wieght,so if we must choose one
#variable we would use the one with the strongest correlation to what we are looking for
#More complex, and likely better, linear regression alogorithms would use more features

X = Fish[['X-Length']]
y = Fish['Weight']


#Step 4: Diviing the dataset into test and train data 
#We are dividing the DataFrame (Object?) into training and test sets
#As it sounds, the training dataset trains the model, the test dataset evalutes its performance

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10, shuffle=True)

#So what does all that stuff do?
#First we pass into train_test_split the feature (X) and the label (Y)
#We use test_size = 0.3 to select 70% of the data for training, and the reamaining 30% for testing purposes
#We make split random (which I presume we don't favor any type of data), we probaly should not do this for
#time related data
#We set shuffle = True to ensure that the model is not overfitting to a specifc set of data
```

```python
from sklearn.linear_model import LinearRegression


# Step 5: Selecting the linear gression method from scikit-Learn libreary
#At this pint we are creating the linear regression library
model = LinearRegression().fit(X_train, y_train)

#Step 6: Validation

#The Mean Absolute Error (MAE) can be calculated using the training and testing data
#Its an indicator of the average aboslute difference between the preidated and actual values
#A lower MAE indciates more accurate predicitons
#Other model validation methods exist

from sklearn import metrics
#Evaluating the trained model on training data 
y_prediction = model.predict(X_train)
print("MAE on train data= ", metrics.mean_absolute_error(y_train, y_prediction))

#Evalutating the trained model on test data
#The mean absolute error (MAE) is defined as the average variance between 
#the significant values in the dataset and the projected values in the same dataset.

y_prediction = model.predict(X_test)
print("MAE on test data = ", metrics.mean_absolute_error(y_test, y_prediction))
```

```python
#multivariable model
# Step 3: Seperating the data into features and labels
newX = Fish[['V-Length', 'D-Length', 'X-Length', 'Height', 'Width']]
y = Fish['Weight']

# Step 4: Dividing the dataset into test and train data
X_train, X_test, y_train, y_test = train_test_split(newX, y, test_size=0.3, random_state=10, shuffle=True)

# Step 5: Selecting the linear regression method from the scikit-learn library
model = LinearRegression().fit(X_train, y_train)

# Step 6: Validation
# Evaluating the trained model on training data
y_prediction = model.predict(X_train)
print("MAE on multivariable train data= " , metrics.mean_absolute_error(y_train, y_prediction))
# Evaluating the trained model on test data
y_prediction = model.predict(X_test)
print("MAE on multivariable test data = " , metrics.mean_absolute_error(y_test, y_prediction))
```

```python
#polynomial model
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt


# Step 3: Seperating the data into features and labels
X = Fish[['V-Length', 'D-Length', 'X-Length', 'Height', 'Width']]
y = Fish['Weight']

# Step 4: Generating polynomial features 
Z = PolynomialFeatures(degree=2, include_bias=False).fit_transform(X)
# Dividing the dataset into test and train data
X_train, X_test, y_train, y_test = train_test_split(Z, y, test_size=0.3, random_state=10)

# Step 5: Selecting the linear regression method from the scikit-learn library
model = LinearRegression().fit(X_train, y_train)

# Step 6: Validation
# Evaluating the trained model on training data
y_prediction = model.predict(X_train)
print("MAE on train data= " , metrics.mean_absolute_error(y_train, y_prediction))
# Evaluating our trained model on test data
y_prediction = model.predict(X_test)
print("MAE on test data = " , metrics.mean_absolute_error(y_test, y_prediction))

```
